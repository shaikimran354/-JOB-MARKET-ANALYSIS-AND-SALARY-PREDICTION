# AI Job Data Analysis and Salary Prediction - EDA Notebook

This notebook performs exploratory data analysis on the AI/ML job salaries dataset.

## Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set style for matplotlib
plt.style.use('default')
sns.set_palette("husl")
```

## Load and Explore Dataset

```python
# Load the dataset
df = pd.read_csv('../data/ds_salaries.csv')

# Display basic information
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData Types:")
print(df.dtypes)
print("\nFirst 5 rows:")
df.head()
```

## Data Cleaning

```python
# Check for missing values
print("Missing values:")
print(df.isnull().sum())

# Check for duplicates
print(f"\nDuplicate rows: {df.duplicated().sum()}")

# Basic statistics
print("\nBasic Statistics:")
df.describe()
```

## Data Visualization

```python
# Salary distribution
fig1 = px.histogram(df, x="salary_in_usd", nbins=50, title="Salary Distribution (USD)")
fig1.show()

# Salary vs Experience Level
exp_map = {'EN': 'Entry', 'MI': 'Mid', 'SE': 'Senior', 'EX': 'Executive'}
df_plot = df.copy()
df_plot['experience_level'] = df_plot['experience_level'].map(exp_map)

fig2 = px.box(df_plot, x="experience_level", y="salary_in_usd", title="Salary by Experience Level")
fig2.show()

# Top job titles
top_jobs = df['job_title'].value_counts().head(10)
fig3 = px.bar(x=top_jobs.values, y=top_jobs.index, orientation='h', title="Top 10 Job Titles")
fig3.show()
```
